```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/weight-height.csv')
df
```

Outlier detection and removal using 3 standard deviation

```python
upper_limit = df.Height.mean() + 3*df.Height.std()
upper_limit
```

```python
lower_limit = df.Height.mean() -3*df.Height.std()
lower_limit
```

```python
df[(df.Height>upper_limit) | (df.Height<lower_limit)]   #select * from df where height <= upper_limit and height >= lower_limit
```

```python
#dataset without outlier
df_no_outlier_std_dev = df[(df.Height<=upper_limit) & (df.Height>=lower_limit)]
df_no_outlier_std_dev
```
