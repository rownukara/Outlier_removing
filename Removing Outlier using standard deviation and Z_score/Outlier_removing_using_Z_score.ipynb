```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/weight-height.csv')
df
```

Outlier detection and removal using Z Score

```python
df['zscore'] = ( df.Height - df.Height.mean() ) / df.Height.std()
df
```

```python
df[df['zscore']>3]
```

```python
df[df['zscore']<-3]
```

```python
#list of the outlier
df[(df.zscore<-3) | (df.zscore>3)]
```

```python
df.zscore
```

```python
df_no_outliers = df[(df.zscore>-3) & (df.zscore<3)]
df_no_outliers
```
