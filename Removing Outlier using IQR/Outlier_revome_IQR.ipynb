```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/heights.csv')
df
```

Detect Outlier using IQR

IQR=[Q3-Q1]

Outlier[Q1-1.5*IQR  ,  Q3+1.5*IQR ]

```python
Q1=df.height.quantile(0.25)
Q3=df.height.quantile(0.75)
Q1,Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_limit=Q1-1.5*IQR
highest_limit=Q3+1.5*IQR
lower_limit,highest_limit
```

Here is the Outlier

```python
#finding outlier
out=df[(df.height<lower_limit)|(df.height>highest_limit)]
out
```

```python
#delete outlier
df_no_outlier=df[(df.height>=lower_limit)&(df.height<=highest_limit)]
df_no_outlier
```
